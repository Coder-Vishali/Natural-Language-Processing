```python
from transformers import pipeline
```

As you can see below , a series of steps are performed:

First of all, we import the pipeline API from the transformers library. 
Make sure to install it in the environment where you have a running installation of (recent) TensorFlow or PyTorch too.

Then, we initialize the Machine Translation pipeline for an *English to German* translation task .

Now, we're ready to translate some text. We input the text *Good evening welcome to this webinar! How are you doing today?* into the translation pipeline.
We then print the outcome on screen.

```python

# Init translator
translator = pipeline("translation_en_to_fr")

# Translate text
text = "Good evening welcome to this webinar, How are you doing today?"
translation = translator(text)

# Print translation
print(translation)
```
